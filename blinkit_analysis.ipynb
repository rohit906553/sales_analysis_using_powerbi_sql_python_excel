 data cleaning

```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF':'Low Fat'})
df['Item Fat Content'] = df['Item Fat Content'].replace({'low fat':'Low Fat'})
df['Item Fat Content'] = df['Item Fat Content'].replace({'reg':'Regular'})
```

```python
print(df['Item Fat Content'].unique())
```

kpi's requirements

```python
#total sales
total_sales= df['Sales'].sum()
print(f"Total sales: $ {total_sales:,.0f}")
```

```python
avg_sales = df['Sales'].mean()
print(f"Average sales: $ {avg_sales:,.1f}")
```

```python
no_of_items_sold = df['Sales'].count()
print(f"No of Items Sold: {no_of_items_sold:,.0f}")
```

```python
avg_ratings= df['Rating'].mean()
print(f"Average ratings:  {avg_ratings:,.1f}")
```

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sales_by_fat, labels= sales_by_fat.index, autopct='%.0f%%', startangle=90)
plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
```

```python
Avg_sales_by_fat = df.groupby('Item Fat Content')['Sales'].mean()
plt.pie(Avg_sales_by_fat, labels= Avg_sales_by_fat.index, autopct='%.1f%%', startangle=90)
plt.title('Avg Sales by Fat Content')
plt.axis('equal')
plt.show()
```

```python
no_of_items_sold_by_fat = df.groupby('Item Fat Content')['Sales'].count()
plt.pie(no_of_items_sold_by_fat, labels= no_of_items_sold_by_fat.index, autopct='%.0f%%', startangle=90)
plt.title('no of items sold by Fat Content')
plt.axis('equal')
plt.show()
```

```python
Avg_ratings_by_fat = df.groupby('Item Fat Content')['Rating'].mean()
plt.pie(Avg_ratings_by_fat, labels= Avg_ratings_by_fat.index, autopct='%.2f%%', startangle=90)
plt.title('Avg Ratings by Fat Content')
plt.axis('equal')
plt.show()
```

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_type.index, sales_by_type.values)
plt.xticks(rotation = -90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():,.0f}', ha='center', va='bottom',fontsize =8)
plt.tight_layout()
plt.show()
```

```python
avg_sales_by_type = df.groupby('Item Type')['Sales'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
bars = plt.bar(avg_sales_by_type.index, avg_sales_by_type.values)
plt.xticks(rotation = -90)
plt.xlabel('Item Type')
plt.ylabel('average Sales')
plt.title('average Sales by Item Type')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():,.0f}', ha='center', va='bottom',fontsize =8)
plt.tight_layout()
plt.show()
```

```python
no_of_items_sold_by_type = df.groupby('Item Type')['Sales'].count().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
bars = plt.bar(no_of_items_sold_by_type.index, no_of_items_sold_by_type.values)
plt.xticks(rotation = -90)
plt.xlabel('Item Type')
plt.ylabel('no of items sold')
plt.title('no of items sold by Item Type')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():,.0f}', ha='center', va='bottom',fontsize =8)
plt.tight_layout()
plt.show()
```

```python
avg_ratings_by_type = df.groupby('Item Type')['Rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
bars = plt.bar(avg_ratings_by_type.index, avg_ratings_by_type.values)
plt.xticks(rotation = -90)
plt.xlabel('Item Type')
plt.ylabel('average rating')
plt.title('average rating by Item Type')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():,.2f}', ha='center', va='bottom',fontsize =8)
plt.tight_layout()
plt.show()
```

sales by outlet location

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)
plt.figure(figsize=(8, 5))#smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data = sales_by_location)
plt.title('Sales by Outlet Location Type')
plt.xlabel('Total Sales')       
plt.ylabel('Outlet Location Type')
plt.tight_layout()
plt.show()
```

```python
avg_sales_by_location = df.groupby('Outlet Location Type')['Sales'].mean().reset_index()
avg_sales_by_location = avg_sales_by_location.sort_values('Sales', ascending=False)
plt.figure(figsize=(8, 5))#smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data = avg_sales_by_location)
plt.title('average Sales by Outlet Location Type')
plt.xlabel('average Sales')       
plt.ylabel('Outlet Location Type')
plt.tight_layout()
plt.show()
```

```python
no_of_items_sold_by_location = df.groupby('Outlet Location Type')['Sales'].count().reset_index()
no_of_items_sold_by_location = no_of_items_sold_by_location.sort_values('Sales', ascending=False)
plt.figure(figsize=(8, 5))#smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data = no_of_items_sold_by_location)
plt.title('no of items sold by Outlet Location Type')
plt.xlabel('no of items sold')       
plt.ylabel('Outlet Location Type')
plt.tight_layout()
plt.show()
```

```python
average_rating_by_location = df.groupby('Outlet Location Type')['Rating'].mean().reset_index()
average_rating_by_location = average_rating_by_location.sort_values('Rating', ascending=False)
plt.figure(figsize=(8, 5))#smaller height, enough width
ax = sns.barplot(x='Rating', y='Outlet Location Type', data = average_rating_by_location)
plt.title('average rating by Outlet Location Type')
plt.xlabel('average rating')       
plt.ylabel('Outlet Location Type')
plt.tight_layout()
plt.show()
```
